```
# nearest-neighbour retrieval: given Sanskrit sentence, pick closest English sentence from corpus.

%cd /content/drive/MyDrive/sanskrit_translation_data
!pip install -e .
!pip install scikit-learn

from src.sanskrit_translation.io_utils import read_csv
from src.sanskrit_translation.paths import CORPUS_CLEANED
from src.sanskrit_translation.embeddings import build_tfidf_model

df = read_csv(CORPUS_CLEANED)
vectorizer, X = build_tfidf_model()

from sklearn.neighbors import NearestNeighbors

nn = NearestNeighbors(n_neighbors=1, metric="cosine")
nn.fit(X)

def retrieve_translation(query: str):
    q_vec = vectorizer.transform([query])
    dist, idx = nn.kneighbors(q_vec)
    idx = idx[0][0]
    return df.loc[idx]

retrieve_translation("धर्मक्षेत्रे कुरुक्षेत्रे")
```
